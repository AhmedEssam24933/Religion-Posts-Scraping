```python
# Path to ChromeDriver
chrome_driver_path = r"C:\Users\Ahmed\chromedriver.exe"  # Update this path

# Set up the WebDriver
driver = webdriver.Chrome()

# Read the DataFrame with links
input_file = r"E:\all tweets\2.5+ Million Rows Egyptian Datasets Collection\Religion Posts\religion links.xlsx"  # Replace with your file containing links
df = pd.read_excel(input_file)

# Create an empty DataFrame to store scraped data
scraped_data = []

try:
    for index, row in df.iterrows():
        link = row['Link']
        print(f"Processing: {link}")
        try:
            # Open the link
            driver.get(link)
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.TAG_NAME, "p"))
            )
            
            # Extract all <p> tags
            paragraphs = driver.find_elements(By.TAG_NAME, "p")
            
            # Collect texts
            for p in paragraphs:
                scraped_data.append({'Link': link, 'Text': p.text.strip()})
        
        except Exception as e:
            print(f"Error processing {link}: {e}")
            continue

    # Save to a new Excel file
    output_file = r"E:\all tweets\2.5+ Million Rows Egyptian Datasets Collection\Religion Posts\religion3.xlsx"
    pd.DataFrame(scraped_data).to_excel(output_file, index=False, engine="openpyxl")
    print(f"Scraped data saved to {output_file}")

finally:
    # Close the WebDriver
    driver.quit()
```

```python
# Input Excel file containing the scraped data
input_file = r"E:\all tweets\2.5+ Million Rows Egyptian Datasets Collection\Religion Posts\religion2.xlsx"
output_file = r"E:\all tweets\2.5+ Million Rows Egyptian Datasets Collection\Religion Posts\splited_sentences.xlsx"

# Load the data into a DataFrame
df = pd.read_excel(input_file)

# Create a new list to store the split data
split_data = []

# Iterate through each row in the DataFrame
for _, row in df.iterrows():
    text = str(row['Text'])  # Convert text to string if it's not already
    
    # Split the text by "،" or ".", replacing "،" with "." to standardize
    split_texts = [segment.strip() for segment in text.replace("،", ".").split(".") if segment.strip()]
    
    # Add each split text as a new row
    for segment in split_texts:
        split_data.append({'Text': segment})

# Create a new DataFrame from the split data
split_df = pd.DataFrame(split_data)

# Save the new DataFrame to an Excel file
split_df.to_excel(output_file, index=False, engine="openpyxl")

print(f"Processed data saved to {output_file}")
```

```python
# Folder containing the Excel files to be merged
folder_path = r"E:\all tweets\2.5+ Million Rows Egyptian Datasets Collection\Religion Posts\data"  # Update this with your folder path
output_file = r"E:\all tweets\2.5+ Million Rows Egyptian Datasets Collection\Religion Posts\data\merged_cleaned_data.xlsx"

# List to hold data from all Excel files
all_data = []

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):  # Check for Excel files
        file_path = os.path.join(folder_path, filename)
        
        # Load the Excel file into a DataFrame
        df = pd.read_excel(file_path, engine="openpyxl")
        
        # Rename the column to 'Text' if it's named differently
        # Check if the column name is not 'Text' and rename it
        if df.columns[0] != 'Text':
            df.rename(columns={df.columns[0]: 'Text'}, inplace=True)  # Rename the first column to 'Text'
        
        # Append the DataFrame to the list
        all_data.append(df)

# Concatenate all DataFrames into one
merged_df = pd.concat(all_data, ignore_index=True)

# Remove rows with NaN values in the 'Text' column
merged_df.dropna(subset=['Text'], inplace=True)

# Remove rows with Text that have 5 or fewer words
merged_df = merged_df[merged_df['Text'].apply(lambda x: len(str(x).split()) >= 5)]

# Save the cleaned DataFrame to an Excel file
merged_df.to_excel(output_file, index=False, engine="openpyxl")

print(f"Cleaned and merged data saved to {output_file}")
```
